# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='ME')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='ME')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

### Using Preattentive Attributes to Highlight the Last Three Months

If I wanted the viewer to look at the last three months first, I would make that part of the graph stand out from the rest. One way to do that would be to keep the first nine months in a light gray and use a bold color for the last three months.

I could also make the last three data points larger, use a thicker line, or add labels only to those months. Another option would be to include a short annotation or lightly shade the background behind the final three months to draw the viewer's attention.

I think the best approach would be to combine a couple of these techniques instead of using all of them. A bold color, slightly larger markers, and a short annotation would naturally guide the viewer to the last three months without making the graph feel too busy. That keeps the graph clean while still emphasizing the part of the data I want the audience to notice first.

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    "Drug A": [1, 1, 2, 1, 3, 3, 4, 2, 1, 4, 2, 7, 1, 1],
    "Drug B": [2, 3, 3, 2, 2, 1, 1, 4, 5, 3, 3, 2, 2, 2],
    "Drug C": [3, 4, 6, 8, 4, 6, 8, 10, 4, 7, 4, 3, 3, 4],
    "Drug D": [6, 5, 12, 4, 8, 5, 10, 9, 6, 9, 5, 4, 6, 3],
    "Drug E": [7, 6, 8, 7, 10, 4, 5, 8, 3, 12, 11, 8, 7, 5],
}

countries = [
    "Australia", "Brazil", "Canada", "China", "France", "Germany",
    "India", "Italy", "Mexico", "Russia", "Spain", "Turkey",
    "United Kingdom", "United States"
]

df_rank = pd.DataFrame(data, index=countries)

plt.figure(figsize=(8, 6))

sns.heatmap(
    df_rank,
    annot=True,
    fmt="d",
    cmap="Blues_r",
    cbar=False,
    linewidths=0.5,
    linecolor="white"
)

plt.title("Top 5 Drugs: Country-Level Sales Rank", loc="left", fontsize=14, fontweight="bold")
plt.xlabel("Drug")
plt.ylabel("Country")

plt.show()
```
